### Calculating Mean for Cases in the state of North Carolina

```python
meansPerWeekNCCase = {}
```

```python
for weekNumberNCCase, (startDateNCCase, endDateNCCase) in enumerate(zip(startDatesOfWeeks, endDatesOfWeeks), start=1):
    weekFilteredNCCase = dfNCDataSet[(dfNCDataSet['Date'] <= endDateNCCase) & (dfNCDataSet['Date'] >= startDateNCCase)]
    meanWeekNCCase = (weekFilteredNCCase['Cases'] / 10.000).mean()
    meansPerWeekNCCase[f'Week {weekNumberNCCase}'] = meanWeekNCCase
```

```python
weekNumbersNCCase = list(meansPerWeekNCCase.keys())
meanValuesNCCase = list(meansPerWeekNCCase.values())
```

```python
plt.figure(figsize=(10, 5))
plt.plot(weekNumbersNCCase, meanValuesNCCase, marker='o', linestyle='-')
plt.title('Weekly Means of Cases')
plt.xlabel('Week')
plt.ylabel('Mean Value of Cases')
plt.grid(True)
plt.xticks(rotation=90)
```

### Distribution of frequency of means for North Carolina

```python
plt.hist(meanValuesNCCase, bins=10, edgecolor='k', alpha=1.0)
plt.title('Distribution of Mean of Weekly Cases in North Carolina')
plt.xlabel('Mean Value')
plt.ylabel('Frequency')
plt.grid(False)
```

#### I wanted to see the frequency of weekly mean values and I distributed the mean of weekly cases in North Carolina per 10000 people to get a general understanding of numbers.

### Calculating the range of values

```python
rangeValuesNCCase= np.histogram(meanValuesNCCase, bins=1)
```

```python
rangeNCCase = (rangeValuesNCCase[0], rangeValuesNCCase[-1])
```

```python
print(rangeNCCase)
```

#### The modality of the distribution looks like Gaussian. Statistical information is calculated below.

```python
meanMeanValuesNCCase = np.mean(meanValuesNCCase)
varianceMeanValuesNCCase = np.var(meanValuesNCCase)
skewnessMeanValuesNCCase = skew(meanValuesNCCase)
kurtosisMeanValuesNCCase = kurtosis(meanValuesNCCase)
```

```python
print(f"Mean: {meanMeanValuesNCCase}")
print(f"Variance: {varianceMeanValuesNCCase}")
print(f"Skewness: {skewnessMeanValuesNCCase}")
print(f"Kurtosis: {kurtosisMeanValuesNCCase}")
```
