```python
# I filtered the main dataset to North Carolina since we will be working with data from North Carolina.

dfCovidDataSet = dfCovidDataSet[dfCovidDataSet['State'] == 'NC']
```

```python
dfCovidDataSet.head(15)
```

```python
# I also filtered the presidential elections dataset to North Carolina since we will be working with data from North Carolina.

dfEnrichmentDataSet = dfEnrichmentDataSet[dfEnrichmentDataSet['state'] == 'North Carolina']
```

```python
dfEnrichmentDataSet.head(15)
```

```python
# I narrowed down the date range to a week to see the COVID-19 cases trend at the beginning of the week as well
# as the end of the week.

WeekStartDate = '2023-07-17'
WeekEndDate = '2023-07-23'

WeekData = dfCovidDataSet[(dfCovidDataSet['Date'] >= WeekStartDate) & (dfCovidDataSet['Date'] <= WeekEndDate)]

WeekData.head(15)
```

```python
# I displayed all the counties' total COVID-19 cases.

plt.figure(figsize=(12, 6))
plt.bar(WeekData["County Name"], WeekData["Cases"])
plt.xlabel("County")
plt.ylabel("Total Cases")
plt.title("Total COVID-19 Cases by County")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# I compared number of cases day by day for each county. According to my analysis, the number of COVID-19
# cases are stable.

plt.figure(figsize=(12, 6))
plt.plot(WeekData['Date'], WeekData['Cases'], marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Daily Cases')
plt.title('Daily COVID-19 Cases for the Week')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Merging the two datasets.

dfMerged = pd.concat([dfCovidDataSet, dfEnrichmentDataSet], axis = 1)
```
