```python
import pandas as pd 
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt



data = pd.read_csv(r"C:\Users\mawul\Downloads\ProjectDataStage1.csv") # reading the csv file 
data 
```

```python
data1 = data.copy() #making a copy 
data1
```

```python
data2 = pd.DataFrame(data1)  #putting it into a dataframe 

data2




```

```python
columns_to_drop = [col for col in data2.columns if (('2020-01-22' <= col <= '2020-07-05') or ('2020-12-28' <= col)) and
                                                  (('Cases' in col) or ('Deaths' in col))]


data2.drop(columns=columns_to_drop, inplace=True)  #dropping columns of unwanted dates 
```

```python
data2
```

```python
data2[data2['State'].str.contains('MA')]  #filtering out massachusetts
```

```python
cases_df = data2.filter(like='Cases', axis=1)  #only case columns 
deaths_df = data2.filter(like='Deaths', axis=1)   #only death columns
```

```python
cases_df
```

```python
cases_df = cases_df.sum(axis=0).to_frame().T.copy(deep=True)

cases_df

```

```python
cases_array = cases_df.to_numpy()

# this is to calculate the sum of the cases in the array using a for loop 
num_columns = cases_array.shape[1]
window_size = 7
sum = []

for i in range(0, num_columns, window_size):
    window = cases_array[:, i:i+window_size]
    window_sum = np.sum(window)
    sum.append(window_sum)

# here is for storing the sum result  
cases_sum_df = pd.DataFrame({'Sum': sum})

# the sum result is printed together with the number of weeks
for week_number, sum in enumerate(sum, start=1):
    print(f"The sum number of cases for week {week_number} was: {sum:.0f}")
```

```python
deaths_df = deaths_df.sum(axis=0).to_frame().T.copy(deep=True)

deaths_df   #this sums up the daily deaths columns 
```

```python
death_array = deaths_df.to_numpy()

# summation of deaths in an array format using for loop
num_columns = death_array.shape[1]
window_size = 7
sum = []

for i in range(0, num_columns, window_size):
    window = death_array[:, i:i+window_size]
    window_sum = np.sum(window)
    sum.append(window_sum)

# result storage 
death_sum_df = pd.DataFrame({'Sum': sum})

# final output of sum reslts and week 
for week_number, sum in enumerate(sum, start=1):
    print(f"The sum number of deaths for week {week_number} was: {sum:.0f}")
```

```python
cases_array = cases_df.to_numpy()

# using for loop to calculate the mean 
num_columns = cases_array.shape[1]
window_size = 7
mean = []

for i in range(0, num_columns, window_size):
    window = cases_array[:, i:i+window_size]
    window_mean = np.mean(window)
    mean.append(window_mean)


cases_mean_df = pd.DataFrame({'Mean': mean})

# outputting the mean 
for week_number, mean in enumerate(mean, start=1):
    print(f"The mean number of cases for week {week_number} was: {mean:.0f}")
```

```python
cases_array = cases_df.to_numpy()

# using for loop to calculate the median 
num_columns = cases_array.shape[1]
window_size = 7
median = []

for i in range(0, num_columns, window_size):
    window = cases_array[:, i:i+window_size]
    window_median = np.median(window)
    median.append(window_median)


cases_median_df = pd.DataFrame({'Median': median})

# outputting the median 
for week_number, median in enumerate(median, start=1):
    print(f"The median number of cases for week {week_number} was: {median:.0f}")
```

```python
import pandas as pd 
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt


#using for loop to calculate the mode and adding stats library since mode is not part of numpy 

num_columns = cases_array.shape[1]
window_size = 7
modes = []

for i in range(0, num_columns, window_size):
    window = cases_array[:, i:i+window_size]
    window_mode, _ = stats.mode(window, axis=None, keepdims=True)  
    modes.append(window_mode[0])  

cases_mode_df = pd.DataFrame({'Mode': modes})



# outputting mode results
for week_number, mode in enumerate(modes, start=1):
    print(f"The mode number of cases for week {week_number} was: {mode:.0f}")
    
    
    
    
```
