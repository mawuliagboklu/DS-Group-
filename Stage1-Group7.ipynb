```python
dfCases = pd.read_csv("Data/covid_confirmed_usafacts.csv")
dfDeath = pd.read_csv("Data/covid_deaths_usafacts.csv")
dfPopulation = pd.read_csv("Data/covid_county_population_usafacts.csv")
```

```python
# Changing display size
pd.set_option('display.max_rows', 500) 
```

```python
# Both cases and deaths have headers with exact same names - I am going to give them unique header names before merging the 
# data

dfDeath = dfDeath.add_suffix('_Deaths')
dfCases = dfCases.add_suffix('_Cases')
```

```python
df = pd.concat([dfPopulation, dfCases], axis=1, join="inner")
df = pd.concat([df, dfDeath], axis=1, join="inner")

df.head()
```

```python
# Defining a list of columns to drop

to_drop = ['countyFIPS_Cases',
          'County Name_Cases',
          'State_Cases',
          'StateFIPS_Cases',
          'countyFIPS_Deaths',
          'County Name_Deaths',
          'State_Deaths',
          'StateFIPS_Deaths']

# Dropping redundant columns
df.drop(to_drop, inplace = True, axis = 1)

```
