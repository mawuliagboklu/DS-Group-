## X-Axis - number of days, Y-Axis number of new cases and deaths. Calcluate error using RMSE.(10 points)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import linregress
import seaborn as sns
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr, ttest_ind
```

```python
# this is refactored code from previous stages, updated to satisfy the different range for this stage.

df = pd.read_csv("Data/ProjectDataStage1WIDEFORMAT.csv")
df2 = df.copy()

columns_to_drop = [col for col in df2.columns if (('2020-01-22' <= col < '2020-06-01') or ('2021-01-04' <= col)) and
                                                  (('Cases' in col) or ('Deaths' in col))]

df2.drop(columns=columns_to_drop, inplace=True)

df2.head(5)

states_to_filter = ["VA"]

state_dataframes = {}

for state in states_to_filter:
    state_df = df2[df2["State"] == state].copy()
    state_dataframes[f"df_{state}"] = state_df

# Taking a look at one of the dataframes to see if everything was handled fine by the function
state_dataframes["df_VA"].head()

def calculate_differences(arr):
    differences = np.diff(arr)
    return differences

def process_dataframe(df):
    cases_sum = np.array([])
    deaths_sum = np.array([])

    cases_columns = [col for col in df.columns if "Cases" in col]
    deaths_columns = [col for col in df.columns if "Deaths" in col]

    if cases_columns:
        cases_sum = df[cases_columns].sum().to_numpy()
        cases_sum = calculate_differences(cases_sum)
    if deaths_columns:
        deaths_sum = df[deaths_columns].sum().to_numpy()
        deaths_sum = calculate_differences(deaths_sum)

    population_sum = df["population"].sum()

    return cases_sum, deaths_sum, population_sum

# create new dataframe name for the data

df_va = state_dataframes["df_VA"]

df_va.head()
```

```python
# county name as index as the index
df_va.set_index("County Name", inplace=True)

# cleaning up the df
df_va.drop("Unnamed: 0", axis=1, inplace=True)
df_va.index.name = None

df_va.head()
```

```python
# these arrays contain data concerning our daily state deaths and cases over the right time range
va_cases, va_deaths, va_pop  = process_dataframe(df_va)

print(va_cases)
```

```python
df_cases = pd.DataFrame({'Day': np.arange(len(va_cases)), 'Cases': va_cases})
df_deaths = pd.DataFrame({'Day': np.arange(len(va_deaths)), 'Deaths': va_deaths})
num_elements_cases = len(va_cases)
num_elements_deaths = len(va_deaths)

plt.figure(figsize=(14, 6))

# linear regression: cases
slope_cases, intercept_cases, r_value_cases, p_value_cases, std_err_cases = linregress(df_cases['Day'], df_cases['Cases'])
regression_line_cases = slope_cases * df_cases['Day'] + intercept_cases

# regresion line
plt.subplot(1, 2, 1)
plt.scatter(df_cases['Day'], df_cases['Cases'], label='Daily Cases')
plt.plot(df_cases['Day'], regression_line_cases, color='red', label='Linear Regression (Trend Line)')

# prediction line
forecast_days_cases = np.arange(df_cases['Day'].max() + 1, df_cases['Day'].max() + 8)
forecast_line_cases = slope_cases * forecast_days_cases + intercept_cases
plt.plot(forecast_days_cases, forecast_line_cases, linestyle='--', color='blue', label='Prediction Path (1 Week Forecast)')

# confidence interval
ci_cases = sns.regplot(x=df_cases['Day'], y='Cases', data=df_cases, scatter=False, ci=95, color='gray', label='Confidence Intervals (95%)')

plt.xlabel('Day')
plt.ylabel('Number of Cases')
plt.title('Statewide Virginia Cases Linear')
plt.legend()

# the same steps as above, but for deaths data
slope_deaths, intercept_deaths, r_value_deaths, p_value_deaths, std_err_deaths = linregress(df_deaths['Day'], df_deaths['Deaths'])
regression_line_deaths = slope_deaths * df_deaths['Day'] + intercept_deaths

plt.subplot(1, 2, 2)
plt.scatter(df_deaths['Day'], df_deaths['Deaths'], label='Daily Deaths', color='green')
plt.plot(df_deaths['Day'], regression_line_deaths, color='orange', label='Linear Regression')

forecast_days_deaths = np.arange(df_deaths['Day'].max() + 1, df_deaths['Day'].max() + 8)
forecast_line_deaths = slope_deaths * forecast_days_deaths + intercept_deaths
plt.plot(forecast_days_deaths, forecast_line_deaths, linestyle='--', color='blue', label='Prediction Path (1 Week Forecast)')

ci_deaths = sns.regplot(x=df_deaths['Day'], y='Deaths', data=df_deaths, scatter=False, ci=95, color='gray', label='Confidence Intervals (95%)')

plt.xlabel('Day')
plt.ylabel('Number of Deaths')
plt.title('Statewide Virginia Deaths Linear')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Function to perform polynomial regression
def polynomial_regression(x, y, degree):
    coefficients = np.polyfit(x, y, degree)
    polynomial = np.poly1d(coefficients)
    return polynomial
```

```python
# It looks like cubic always has the lowest value, and is the best representation of the data

# creating a function to calculate slope of a tangent line, which we will use to evaluate the severity
# of the case growth in all counties
def slope_of_tangent_line(polynomial, point):
    derivative_coefficients = np.polyder(polynomial.coefficients)
    slope = np.polyval(derivative_coefficients, point)
    return slope

# code from above adapted to counties
def process_county(df, location_name):
    cases_sum = np.array([])
    deaths_sum = np.array([])

    cases_columns = [col for col in df.columns if "Cases" in col]
    deaths_columns = [col for col in df.columns if "Deaths" in col]

    if location_name in df.index:
        if cases_columns:
            cases_sum = df.loc[location_name, cases_columns].to_numpy()
            cases_sum = calculate_differences(cases_sum)
            cases_sum[cases_sum < 0] = 0
        if deaths_columns:
            deaths_sum = df.loc[location_name, deaths_columns].to_numpy()
            deaths_sum = calculate_differences(deaths_sum)
            deaths_sum[deaths_sum < 0] = 0

    population_sum = df.loc[location_name, "population"] if location_name in df.index else 0

    return cases_sum, deaths_sum, population_sum
```

Here you would compare the covid cases to the state or county level enrichment data to prove or disprove your null hypothesis. In this case there will be a two tail - two sample t-test to see if there is a difference and then one-tail - two sample t-test to show higher or lower.
### you can also make 3 new hypotheses and do hypothesis tests on them.

Hypothesis 1. Does an increase in jobs/employment in an area lead to/correlate with an increased number of COVID cases? 

```python
# A function that can be used to calculate whether significant correlation exists between two arrays.
# This tests the null hypothesis of a correlation coefficient of 0

def correlated_and_t_test(array1, array2, alpha=0.05):
    # pearson correlation
    correlation_coefficient, p_value = pearsonr(array1, array2)

    # chck statistical significance
    if p_value < alpha:
        if correlation_coefficient > 0:
            print("There is a statistically significant positive correlation.")
        elif correlation_coefficient < 0:
            print("There is a statistically significant negative correlation.")
        else:
            print("There is no correlation.")
    else:
        t_stat, p_value_lower = ttest_ind(array1, array2, alternative='less')
        t_stat, p_value_higher = ttest_ind(array1, array2, alternative='greater')

        if p_value_lower < alpha:
            print("There is a statistically significant lower correlation.")
        elif p_value_higher < alpha:
            print("There is a statistically significant higher correlation.")
        else:
            print("There is no statistically significant correlation.")
```

```python
county_names = ['Dinwiddie County', 'Henry County', 'Charles City County', 'Buena Vista City', 'Lancaster County']

county_dfs = {}  

for county_name in county_names:
    county_cases_df = cases_df[cases_df['County Name'] == county_name]
    
    county_array = county_cases_df.filter(like='_Cases').sum(axis=0).to_numpy(dtype=np.int64)
    
    county_data = np.diff(county_array, prepend=0)
    
    start_date = '2020-01-22'
    county_date_range = pd.date_range(start=start_date, periods=len(county_data), freq='D')
    county_data_dict = {'Date': county_date_range, county_name: county_data}
    county_df = pd.DataFrame(county_data_dict)
    
    county_df['Date'] = county_df['Date'].astype(str)
    
    county_dfs[county_name] = county_df

sum_cases_counties = pd.concat([df.set_index('Date') for df in county_dfs.values()], axis=1).reset_index()

sum_cases_counties = sum_cases_counties.transpose()

sum_cases_counties.columns = sum_cases_counties.iloc[0]
sum_cases_counties = sum_cases_counties.iloc[1:]

sum_cases_counties.head(8)
```

```python
sum_cases_counties.columns = pd.to_datetime(sum_cases_counties.columns)

start_date = '2023-01-01'
end_date = '2023-03-31'

filtered_columns = sum_cases_counties.columns[(sum_cases_counties.columns >= start_date) & (sum_cases_counties.columns <= end_date)]

sum_cases_counties_filtered = sum_cases_counties[filtered_columns]

sum_cases_counties_filtered = sum_cases_counties_filtered.clip(lower=0)

sum_cases_counties_filtered.head(8)
```

```python
arrays = {}

for index, row in sum_cases_counties_filtered.iterrows():
    sum_january = row.filter(like='2023-01-').sum()
    sum_february = row.filter(like='2023-02-').sum()
    sum_march = row.filter(like='2023-03-').sum()

    array_name = f"{index.lower().replace(' ', '_')}_sums"
    arrays[array_name] = np.array([sum_january, sum_february, sum_march])

for array_name, array in arrays.items():
    print(f"{array_name}: {array}")
```

```python
print("In Dinwiddie County:")
correlated_and_t_test(arrays['dinwiddie_county_sums'], dinw_emp, alpha=0.05)
print("In Henry County:")
correlated_and_t_test(arrays['henry_county_sums'], henry_emp, alpha=0.05)
print("In Charles City County:")
correlated_and_t_test(arrays['charles_city_county_sums'], charles_emp, alpha=0.05)
print("In Buena Vista city:")
correlated_and_t_test(arrays['buena_vista_city_sums'], bvc_emp, alpha=0.05)
print("In Lancaster County:")
correlated_and_t_test(arrays['lancaster_county_sums'], lancaster_emp, alpha=0.05)
```

Because 4 out of 5 counties displayed a significant correlation that is lower than 0, I would say that the null hypothesis is true, but my initial, alternate hypothesis, that an increase in jobs would lead to an increase in cases, is not supported, and there is in fact a negative relation between the two datasets. This matches the data I compared between states in Stage 3.

Hypothesis 2: Does an increase in Republican Voters in a county increase the amount of COVID cases there?

```python
# loading the csv containing election data
electiondf = pd.read_csv("Data/president_county_candidate.csv")
electiondf.head()
```

```python
counties_to_keep = ['Dinwiddie County', 'Henry County', 'Charles City County', 'Buena Vista city', 'Lancaster County']
parties_to_keep = ['REP', 'DEM']

electiondf = electiondf[
    (electiondf['county'].str.contains('|'.join(counties_to_keep), case=False, na=False)) & 
    (electiondf['party'].isin(parties_to_keep)) &
    (electiondf['state'] == 'Virginia')
]

electiondf.head(50)
```

```python
sums_dict = {}

for county in counties_to_keep:
    for party in parties_to_keep:
        key = f'{county} {party}'
        
        total_votes_sum = electiondf[(electiondf['county'] == county) & (electiondf['party'] == party)]['total_votes'].sum()
        
        sums_dict[key] = total_votes_sum

print(sums_dict)
```

```python
dinw_vote = sums_dict['Dinwiddie County REP'] / (sums_dict['Dinwiddie County REP'] + sums_dict['Dinwiddie County DEM'])
henry_vote = sums_dict['Henry County REP'] / (sums_dict['Henry County REP'] + sums_dict['Henry County DEM'])
charles_vote = sums_dict['Charles City County REP'] / (sums_dict['Charles City County REP'] + sums_dict['Charles City County DEM'])
bvc_vote = sums_dict['Buena Vista city REP'] / (sums_dict['Buena Vista city REP'] + sums_dict['Buena Vista city DEM'])
lancaster_vote = sums_dict['Lancaster County REP'] / (sums_dict['Lancaster County REP'] + sums_dict['Lancaster County DEM'])
print('The percent of people in Dinwiddie that voted Republican is: ',f"{dinw_vote * 100:.2f}%")
```

```python
# placing vote data and case data in arrays where case data and vote data are at the same index
county_cases_averages = np.array([
    np.mean(henry_cases),
    np.mean(dinw_cases),
    np.mean(charles_cases),
    np.mean(lancaster_cases),
    np.mean(bvc_cases)
])

county_republican_votes = np.array([
    henry_vote,
    dinw_vote,
    charles_vote,
    lancaster_vote,
    bvc_vote
])

print(county_cases_averages)
print(county_republican_votes)
```

```python
correlated_and_t_test(county_cases_averages, county_republican_votes, alpha=0.05)
```

At least in the counties that I tested, there is no statistically significant correlation between the voting preference and the amount of cases in these counties. This lines up with what I discovered when testing this hypothesis among states, as well, in Stage 3.
