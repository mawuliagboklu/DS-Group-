### Adding the column for the number of days to our dataframe

```python
dfCovidDataSet['Date'] = pd.to_datetime(dfCovidDataSet['Date'])
```

```python
dfCovidDataSet['numberOfDays'] = (dfCovidDataSet['Date'] - pd.to_datetime('2020-06-01')).dt.days + 1
```

### Putting values for the same counties together

```python
dfCovidDataSet.sort_values(by=['County Name', 'Date'], inplace=True)
```

### Finding the exact daily values

```python
dfCovidDataSet['temporaryDiffOfCases'] = dfCovidDataSet.groupby('County Name')['Cases'].diff()
dfCovidDataSet['temporaryDiffOfDeaths'] = dfCovidDataSet.groupby('County Name')['Deaths'].diff()
```

```python
dfCovidDataSet.head()
```

```python
dfCovidDataSet['Cases'] = dfCovidDataSet['temporaryDiffOfCases'].fillna(dfCovidDataSet['Cases'])
dfCovidDataSet['Deaths'] = dfCovidDataSet['temporaryDiffOfDeaths'].fillna(dfCovidDataSet['Deaths'])
```

```python
dfCovidDataSet.head()
```

```python
dfCovidDataSet.drop(['temporaryDiffOfCases', 'temporaryDiffOfDeaths'], axis=1, inplace=True)
```

### Displaying the number of cases per day

### Started working on Regressions by importing the libraries

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

### North Carolina Cases Linear Regression

```python
ncTotalCasesX = np.array(dfNCDailyCases.numberOfDays).reshape(-1,1)
ncTotalCasesY = np.array(dfNCDailyCases.Cases).reshape(-1,1)
```

```python
ncTotalCasesModelLin = LinearRegression()
ncTotalCasesModelLin.fit(ncTotalCasesX, ncTotalCasesY)
ncTotalCasesModelPredLin = ncTotalCasesModelLin.predict(ncTotalCasesX)
```

```python
plt.figure(figsize=(9, 5))
plt.scatter(dfNCDailyCases['numberOfDays'], dfNCDailyCases['Cases'], label='Number of Cases')
plt.plot(dfNCDailyCases['numberOfDays'], ncTotalCasesModelPredLin, color='red', label='Linear Regression Line')
plt.title('Linear Regression for North Carolina Cases')
plt.xlabel('Number of Days')
plt.ylabel('Number of Cases')
plt.legend()
plt.show()
```

### North Carolina Cases Polynomial Regression

```python
nValue = 3
ncTotalCasesModelPolyFeatures = PolynomialFeatures(degree = nValue)
ncTotalCasesXPoly = ncTotalCasesModelPolyFeatures.fit_transform(ncTotalCasesX)
ncTotalCasesModelPoly = LinearRegression()
ncTotalCasesModelPoly.fit(ncTotalCasesXPoly, ncTotalCasesY)
ncTotalCasesModelPredPoly = ncTotalCasesModelPoly.predict(ncTotalCasesXPoly)
```

```python
plt.figure(figsize=(9, 5))
plt.scatter(dfNCDailyCases['numberOfDays'], ncTotalCasesY, label='Actual Cases')
plt.plot(dfNCDailyCases['numberOfDays'], ncTotalCasesModelPredPoly, color='green', label=f'Polynomial Regression Line')
plt.title('Polynomial Regression for North Carolina Cases')
plt.xlabel('Number of Days')
plt.ylabel('Number of Cases')
plt.legend()
```

### North Carolina Deaths Linear Regression

```python
ncTotalDeathsX = np.array(dfNCDailyDeaths.numberOfDays).reshape(-1,1)
ncTotalDeathsY = np.array(dfNCDailyDeaths.Deaths).reshape(-1,1)
```

```python
ncTotalDeathsModelLin = LinearRegression()
ncTotalDeathsModelLin.fit(ncTotalDeathsX, ncTotalDeathsY)
ncTotalDeathsModelPredLin = ncTotalDeathsModelLin.predict(ncTotalDeathsX)
```

```python
plt.figure(figsize=(9, 5))
plt.scatter(dfNCDailyDeaths['numberOfDays'], dfNCDailyDeaths['Deaths'], label='Number of Deaths')
plt.plot(dfNCDailyDeaths['numberOfDays'], ncTotalDeathsModelPredLin, color='red', label='Linear Regression Line')
plt.title('Linear Regression for North Carolina Deaths')
plt.xlabel('Number of Days')
plt.ylabel('Number of Deaths')
plt.legend()
plt.show()
```

### North Carolina Deaths Polynomial Regression

```python
nValue = 3
ncTotalDeathsModelPolyFeatures = PolynomialFeatures(degree = nValue)
ncTotalDeathsXPoly = ncTotalDeathsModelPolyFeatures.fit_transform(ncTotalDeathsX)
ncTotalDeathsModelPoly = LinearRegression()
ncTotalDeathsModelPoly.fit(ncTotalDeathsXPoly, ncTotalDeathsY)
ncTotalDeathsModelPredPoly = ncTotalDeathsModelPoly.predict(ncTotalDeathsXPoly)
```

```python
plt.figure(figsize=(9, 5))
plt.scatter(dfNCDailyDeaths['numberOfDays'], ncTotalDeathsY, label='Actual Cases')
plt.plot(dfNCDailyDeaths['numberOfDays'], ncTotalDeathsModelPredPoly, color='green', label=f'Polynomial Regression Line')
plt.title('Polynomial Regression for North Carolina Deaths')
plt.xlabel('Number of Days')
plt.ylabel('Number of Deaths')
plt.legend()
```

### RMSE Values for all 4 Modelings

```python
rmseNCCasesLin = mean_squared_error(ncTotalCasesY, ncTotalCasesModelPredLin, squared=False)
rmseNCCasesPoly = mean_squared_error(ncTotalCasesY, ncTotalCasesModelPredPoly, squared=False)
rmseNCDeathsLin = mean_squared_error(ncTotalDeathsY, ncTotalDeathsModelPredLin, squared=False)
rmseNCDeathsPoly = mean_squared_error(ncTotalDeathsY, ncTotalDeathsModelPredPoly, squared=False)
```

```python
print(f"RMSE for North Carolina Cases - Linear Regression: {rmseNCCasesLin}")
print(f"RMSE for North Carolina Cases - Polynomial Regression: {rmseNCCasesPoly}")
print(f"RMSE for North Carolina Deaths - Linear Regression: {rmseNCDeathsLin}")
print(f"RMSE for North Carolina Deaths - Polynomial Regression: {rmseNCDeathsPoly}")
```
