Long Format Data 

The following code creates the same dataframe but in LONG FORMAT

```python
# Resetting my dataframes, to remove the _cases and _deaths I added to the column headers

dfCasesWide = pd.read_csv("Data/covid_confirmed_usafacts.csv")
dfDeathWide = pd.read_csv("Data/covid_deaths_usafacts.csv")
```

```python
# Creating a list of dates that are used as column headers in the death and cases data, 
# so I can put the table in long format.  I originally just made a list of every date
# between 2020/01/22 and 2023/07/23, but that caused errors when I got the the melt 
# step, since there is not data for every single date within that range.

dates = dfCasesWide.columns.tolist()

print(dates)

# removing the first four values in the list using pop
dates.pop(0)
dates.pop(0)
dates.pop(0)
dates.pop(0)

```

```python
# Reshape from wide to long format using melt and the list of column headers created above
# from pandas.org:  DataFrame.melt(id_vars=None, value_vars=None, var_name=None, value_name='value', col_level=None, ignore_index=True)[source]

dfCasesLong = pd.melt(dfCasesWide, id_vars=['countyFIPS', 'County Name', 'State', 'StateFIPS'], value_vars=dates, var_name = "Date", value_name='Cases')
dfCasesLong.head(50)

```

```python
# Repeating same steps with the dfDeathsWide

dfDeathsLong = pd.melt(dfDeathWide, id_vars=['countyFIPS', 'County Name', 'State', 'StateFIPS'], value_vars=dates, var_name = "Date", value_name='Deaths' )
dfDeathsLong.head(20)
```

```python
# Before concatenating the tables togther, we need to create a unique ID to combine them on.  It will have to be a combo 
# of county and date.

# bypassed code, it's unnecessary

#dfCasesLong['countyFIPS'] = dfCasesLong['countyFIPS'].apply(str)
#dfDeathsLong['countyFIPS'] = dfDeathsLong['countyFIPS'].apply(str)

#dfCasesLong['rowID'] = dfCasesLong['countyFIPS'] + dfCasesLong['Date']
#dfDeathsLong['rowID'] = dfDeathsLong['countyFIPS'] + dfDeathsLong['Date']

#dfDeathsLong.head()
```

```python
## Merging cases and deaths dataframes

dfLong = pd.merge(dfCasesLong, dfDeathsLong,  left_index=True, right_index=True)
dfLong = dfLong.loc[:, ~dfLong.columns.duplicated()]

# dropping duplicate columns
list = ['countyFIPS_x',
        'County Name_x',
        'State_x',
        'StateFIPS_x',
        'Date_x']
dfLong.drop(list, inplace = True, axis = 1)


# renaming remaining columns

dfLong.rename(columns={"countyFIPS_y": "countyFIPS"}, inplace=True)
dfLong.rename(columns={"County Name_y": "County Name"}, inplace=True)
dfLong.rename(columns={"StateFIPS_y": "StateFIPS"}, inplace=True)
dfLong.rename(columns={"Date_y": "Date"}, inplace=True)
dfLong.rename(columns={"State_y": "State"}, inplace=True)




dfLong.head()
```

```python
#  The join appears to work correctly

print(dfCasesLong.shape)
print(dfDeathsLong.shape)
print(dfLong.shape)
print(dfLong.isnull().sum())
```

```python
# Outputing prepared data frame

dfLong.to_csv('ProjectDataStage1LONGFORMAT.csv')
```
